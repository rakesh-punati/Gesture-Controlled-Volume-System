```python
import pyautogui as pg
import cv2
import mediapipe as mp
import math
import numpy as np
import time 

hand = mp.solutions.hands.Hands()

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=1,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

mp_draw = mp.solutions.drawing_utils

prev_distance = None

video = cv2.VideoCapture(0)

pTime = 0
while video.isOpened():
    success, Frame = video.read()
    if not success:
        break

    rgb_image = cv2.cvtColor(Frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_image)

    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime

    cv2.putText(Frame, f"Fps:{int(fps)}",
                (50,100),
                cv2.FONT_HERSHEY_SIMPLEX, 1.5,
                (0, 0, 255), 3)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_draw.draw_landmarks(
                Frame,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS
            )

            h, w, _ = Frame.shape
        
            thumb_tip = hand_landmarks.landmark[4]
            index_tip = hand_landmarks.landmark[8]
                    
            x1, y1 = int(thumb_tip.x * w), int(thumb_tip.y * h)
            x2, y2 = int(index_tip.x * w), int(index_tip.y * h)
                    
            distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5

            cx = (x1+x2)//2
            cy = (y1+y2)//2
            
            cv2.circle(Frame, (x1, y1), 8, (255,0,0), -1)
            cv2.circle(Frame, (x2, y2), 8, (255,0,0), -1)
            cv2.line(Frame, (x1,y1),(x2,y2),(255,0,0), 3)
            cv2.circle(Frame,(cx,cy),8,(255,0,0),-1)

            cv2.putText(Frame, f"dist:{int(distance)}",
                (cx+20,cy-20),
                cv2.FONT_HERSHEY_SIMPLEX, 1.5,
                (255, 0, 255), 3)

            if distance < 30:
                cv2.circle(Frame,(cx,cy),8,(0,0,255,),-1)

            if distance > 170:
                cv2.circle(Frame,(cx,cy),8,(0,255,0),-1)
            
            if prev_distance is not None:
                if distance > prev_distance+3:   # small buffer to avoid shaking
                    pg.press("volumeup",presses=int(0.4*(distance-prev_distance)))
                elif distance < prev_distance+3:
                    pg.press("volumedown",presses=int(0.4*(prev_distance-distance)))

            prev_distance = distance
            
        
    cv2.imshow("video", Frame)
    if cv2.waitKey(1) & 255 == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
